## Exercise 3 and HW2 #2

The CN method is used to estimate direct runoff or infiltration from rainfall events in a particular watershed. The CN value is based on the area's land use, soil type, and antecedent moisture condition, effectively acting as a "weight" that influences the calculation of runoff from precipitation.

### Urban Basin:
- **Location**: At the northern edge, directly upstream of the central basin.
- **Characteristics**: Highly urbanized area with impervious surfaces like concrete and asphalt.
- **CN Value**: High (around 85-98) due to low infiltration rates and high runoff potential.

### Agricultural Basin:
- **Location**: Eastern side, connected to the central basin via a small stream.
- **Characteristics**: Predominantly agricultural lands with mixed crops and some areas of bare soil.
- **CN Value**: Moderate to high (around 65-85) depending on soil type, crop cover, and tillage practices.

### Forested Basin:
- **Location**: To the west, with a natural watercourse flowing into the central basin.
- **Characteristics**: Dense forest cover with well-developed soil organic layers.
- **CN Value**: Lower (around 30-55) due to high infiltration rates and low runoff potential.

Suppose, we can calucate the potential runoff (PR) with CN and rainfall data as:

### PR = CN * basin rainfall (kg/day)
